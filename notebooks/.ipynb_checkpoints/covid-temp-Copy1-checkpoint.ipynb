# Conduct Data Cleansing

```python
# Homework

# rename kolom (df.rename)
# SNo --> id
# ObservationDate --> date_observed
# Province/State --> province
# Country/Region --> country
# Last Update --> date_updated
# Confirmed --> n_confirm
# Deaths --> n_death
# Recovered --> n_recover

# hapus karakter yg tidak sesuai dari country (bukan hapus row nya)
# ('St. Martin',) --> St. Martin
```

```python
df_clean = df.copy()
```

```python
# rename columns

df_clean = df_clean.rename(columns={"SNo": "id",
                                    "ObservationDate": "date_observed",
                                    "Province/State": "province",
                                    "Country/Region": "country",
                                    "Last Update": "date_updated",
                                    "Confirmed": "n_confirm",
                                    "Deaths": "n_death",
                                    "Recovered": "n_recover"
                                   }
                          )

df_clean.sample(10)
```

```python
df_clean[df_clean["country"] == "('St. Martin',)"]
```

```python
df_clean[df_clean["country"].str.contains(",")]
```

```python
df_clean["country"].str.contains(",")
```

```python
# replace ('St. Martin',) --> St. Martin in country

df_clean.loc[4770, "country"] = "St. Martin"
```

```python
df_clean.loc[4770]
```

```python
# move The to the front of , detected countries

df_clean[df_clean["country"].str.contains(",")]
```

```python
df_clean.loc[6807]["country"].replace("Gambia, The", "Gambia")
```

```python
df_clean.loc[df_clean["country"].str.contains(","), "country"] = df_clean["country"].str.replace("Gambia, The", "Gambia")
df_clean.loc[df_clean["country"].str.contains(","), "country"] = df_clean["country"].str.replace("Bahamas, The", "Bahamas")
```

```python
df_clean[df_clean["country"].str.contains("\\)")]["country"].value_counts()
```

```python
# convert date_observed to datetime

df_clean
```

```python
df_clean.info()
```

```python
df_clean["date_observed"].str.len().value_counts()
```

```python
df_clean
```

```python
df_clean["date_observed2"] = pd.to_datetime(df_clean["date_observed"])
```

```python
df_clean
```

```python
df_clean.info()
```

```python
df_clean[["date_observed", "date_observed2"]].sample(10)
```

```python
# convert date_updated to datetime

df_clean["date_updated"].str.len().value_counts()
```

```python
df_clean["date_updated2"] = pd.to_datetime(df_clean["date_updated"], format="mixed")
```

```python
df_clean
```

```python
# add 7 hours to date_updated

df_clean["date_updated3"] = df_clean["date_updated2"] + pd.Timedelta(hours=7)
```

```python
df_clean
```

```python
df_clean.info()
```

```python
df_clean[["date_observed", "date_observed2", "date_updated", "date_updated2", "date_updated3"]].sample(10)
```

```python
# drop the old date columns

df_clean = df_clean.drop(["date_observed", "date_updated", "date_updated2"], axis=1)
```

```python
df_clean = df_clean.rename(columns={"date_observed2": "date_observed", "date_updated3": "date_updated"})
```

```python
df_clean
```

```python
# add a new column date_gap_hour

df_clean["date_gap_hour"] = round((df_clean["date_updated"] - df_clean["date_observed"]).dt.total_seconds()/3600, 2).abs()

df_clean
```

```python
df_clean.info()
```

```python
# convert float to int

for i in ["n_confirm", "n_death", "n_recover"]:
    df_clean[i] = df_clean[i].astype("int")

df_clean
```

```python
# reorder

df_clean = df_clean[["id", "province", "country", "date_observed", "date_updated", "date_gap_hour", "n_confirm", "n_death", "n_recover"]]

df_clean
```

```python
# save to csv/excel

df_clean.to_csv("../data/covid_19_data_cleaned.csv", index=False)
df_clean.to_excel("../data/covid_19_data_cleaned.xlsx", index=False)
```

## Further Perform EDA and Cleansing

```python
df2 = pd.read_csv("../data/covid_19_data_cleaned.csv")
for i in ["date_observed", "date_updated"]:
    df2[i] = pd.to_datetime(df2[i], format="ISO8601")

df2
```

### Notes
- sifat dari `n_confirm`, `n_death`, dan `n_recover` adalah cumulative
- bagaimana cara mendapatkan delta masing-masing hari

### Possible Goal
- 

### Clean Action
- dapatkan delta per hari

```python
df3 = df2.copy()
df3["province"] = df3["province"].fillna("Unknown")
df3 = df3.drop(["id", "date_updated", "date_gap_hour"], axis=1)

country = None
province = None
if country:
    df3 = df3[df3["country"] == country]
    if province:
        df3 = df3[df3["province"] == province]
```

```python
# delta_col = ["confirm", "death"]

# for col in delta_col:
#     df3[f"delta_{col}"] = df3.groupby(["country", "province"])[f"n_{col}"].diff().fillna(df3[f"n_{col}"])
#     df3[f"delta_{col}"] = df3[f"delta_{col}"].astype("int")

vis0 = df3.copy()

vis0_max_date = vis0.copy()
vis0_max_date = vis0[["province", "country", "date_observed"]]
vis0_max_date = vis0_max_date.groupby(["country", "province"]).max().reset_index()
vis0_max_date["key"] = vis0_max_date["country"].str.strip() + "_" + vis0_max_date["province"].str.strip() + "_" + vis0_max_date["date_observed"].astype("str").str.replace("-", "")
vis0_max_date = vis0_max_date.drop(["country", "province", "date_observed"], axis=1)

vis0_with_key = vis0.copy()
vis0_with_key["key"] = vis0_with_key["country"].str.strip() + "_" + vis0_with_key["province"].str.strip() + "_" + vis0_with_key["date_observed"].astype("str").str.replace("-", "")

vis0_final = pd.merge(vis0_max_date, vis0_with_key, how="left", on=["key"])
# vis0_final = vis0_final[vis0_final["key"].notna()]
vis0_final = vis0_final[["country", "n_confirm", "n_death", "n_recover"]]
vis0_final = vis0_final.groupby("country").sum().reset_index()
vis0_final = vis0_final.sort_values("n_confirm", ascending=False)
vis0_final = vis0_final.head(5)
vis0_final = vis0_final.melt(id_vars="country", value_vars=["n_confirm", "n_recover", "n_death"], var_name="measurement", value_name="total")
# vis0_final = vis0_final.sort_values("total")

# vis0_with_key[(vis0_with_key["country"] == "Russia") & (vis0_with_key["province"] == "Yaroslavl Oblast")]
# vis0_max_date[(vis0_max_date["country"] == "Russia") & (vis0_max_date["province"] == "Yaroslavl Oblast")]

vis0_final
```

```python
vis0_final["country"].value_counts()
```

```python
fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(20, 10))
ax = ax.flatten()

sns.barplot(ax=ax[0], data=vis0_final, x="country", y="total", hue="measurement")
```

```python
vis1 = df3.copy()

vis1_max_date = vis1.copy()
vis1_max_date = vis1_max_date[["province", "country", "date_observed"]]
vis1_max_date = vis1_max_date.groupby(["country", "province"]).max().reset_index()
vis1_max_date["key"] = vis1_max_date["country"].str.strip() + "_" + vis1_max_date["province"].str.strip() + "_" + vis1_max_date["date_observed"].astype("str").str.replace("-", "")
vis1_max_date = vis1_max_date.drop(["country", "province", "date_observed"], axis=1)

vis1_with_key = vis1.copy()
vis1_with_key["key"] = vis1_with_key["country"].str.strip() + "_" + vis1_with_key["province"].str.strip() + "_" + vis1_with_key["date_observed"].astype("str").str.replace("-", "")

vis1_final = vis1_max_date.merge(vis1_with_key, how="left", on=["key"])
vis1_final = vis1_final[vis1_final["key"].notna()]
vis1_final = vis1_final[["country", "province", "n_confirm", "n_death", "n_recover"]]
# vis1_final["country_province"] = vis1_final["country"] + "_" + vis1_final["province"]
vis1_final = vis1_final.groupby(["country", "province"]).sum().reset_index()
vis1_final = vis1_final.sort_values("n_confirm", ascending=False)
vis1_final = vis1_final.head(5)
vis1_final = vis1_final.melt(id_vars=["country", "province"], value_vars=["n_confirm", "n_recover", "n_death"], var_name="measurement", value_name="total")

vis1_final
```

```python
# delta_col = ["confirm", "death"]

# for col in delta_col:
#     df3[f"delta_{col}"] = df3.groupby(["country", "province"])[f"n_{col}"].diff().fillna(df3[f"n_{col}"])
#     df3[f"delta_{col}"] = df3[f"delta_{col}"].astype("int")

vis3 = df3.copy()

vis3_max_date = vis3.copy()
vis3_max_date = vis3[["province", "country", "date_observed"]]
vis3_max_date = vis3_max_date.groupby(["country", "province"]).max().reset_index()
vis3_max_date["key"] = vis3_max_date["country"].str.strip() + "_" + vis3_max_date["province"].str.strip() + "_" + vis3_max_date["date_observed"].astype("str").str.replace("-", "")
vis3_max_date = vis3_max_date.drop(["country", "province", "date_observed"], axis=1)

vis3_with_key = vis3.copy()
vis3_with_key["key"] = vis3_with_key["country"].str.strip() + "_" + vis3_with_key["province"].str.strip() + "_" + vis3_with_key["date_observed"].astype("str").str.replace("-", "")

vis3_final = vis3_max_date.merge(vis3_with_key, how="left", on=["key"])
vis3_final = vis3_final[vis3_final["key"].notna()]
vis3_final = vis3_final[["country", "n_confirm", "n_death", "n_recover"]]
vis3_final = vis3_final.groupby("country").sum().reset_index()
vis3_final = vis3_final.sort_values("n_confirm", ascending=False)
vis3_final = vis3_final.head(5)
vis3_final = vis3_final.melt(id_vars="country", value_vars=["n_confirm", "n_recover", "n_death"], var_name="measurement", value_name="total")

vis3_final
```

```python
vis4 = df3.copy()

vis4_max_date = vis4.copy()
vis4_max_date = vis4_max_date[["province", "country", "date_observed"]]
vis4_max_date = vis4_max_date.groupby(["country", "province"]).max().reset_index()
vis4_max_date["key"] = vis4_max_date["country"].str.strip() + "_" + vis4_max_date["province"].str.strip() + "_" + vis4_max_date["date_observed"].astype("str").str.replace("-", "")
vis4_max_date = vis4_max_date.drop(["country", "province", "date_observed"], axis=1)

vis4_with_key = vis4.copy()
vis4_with_key["key"] = vis4_with_key["country"].str.strip() + "_" + vis4_with_key["province"].str.strip() + "_" + vis4_with_key["date_observed"].astype("str").str.replace("-", "")

vis4_final = vis4_max_date.merge(vis4_with_key, how="left", on=["key"])
vis4_final = vis4_final[vis4_final["key"].notna()]
vis4_final = vis4_final[["country", "province", "n_confirm", "n_death", "n_recover"]]
# vis4_final["country_province"] = vis4_final["country"] + "_" + vis4_final["province"]
vis4_final = vis4_final.groupby(["country", "province"]).sum().reset_index()
vis4_final = vis4_final.sort_values("n_confirm", ascending=False)
vis4_final = vis4_final.head(5)
vis4_final = vis4_final.melt(id_vars=["country", "province"], value_vars=["n_confirm", "n_recover", "n_death"], var_name="measurement", value_name="total")

vis4_final
```

```python
fig, ax = plt.subplots(ncols=4, nrows=2, figsize=(25, 15))
ax = ax.flatten()

sns.barplot(ax=ax[0], data=vis0_final, x="country", y="total", hue="measurement")
sns.barplot(ax=ax[1], data=vis1_final, x="province", y="total", hue="measurement")

sns.barplot(ax=ax[2], data=vis3_final, x="country", y="total", hue="measurement")
sns.barplot(ax=ax[3], data=vis4_final, x="province", y="total", hue="measurement")
```
