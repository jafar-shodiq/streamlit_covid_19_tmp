```python
import pandas as pd

df = pd.read_csv("../data/children_thr_data.csv")
df["age"] = df["age"].astype("int")

def distribute_thr(total_budget):
    """
    Calculate Tunjangan Hari Raya (THR) to be distributed for each of the children (given dataframe)

    total_budget: int: initial THR budget
    """
    thr = total_budget
    
    for name,age in zip(df["name"], df["age"]):
        if age >= 1 and age <= 5:
            thr = 20000
        elif age >= 6 and age <= 10:
            thr = 30000
        elif age >= 11 and age <= 15:
            thr = 50000
        elif age >= 16 and age <= 20:
            thr = 75000
        elif age >= 21 and age <= 25:
            thr = 100000
        else:
            thr = 0

        if thr == 0:
            print(f"{name}-{age}thn tidak dapat sama sekali")
            continue
        elif total_budget-thr < 0:
            print(f"{name}-{age}thn. harusnya dapat {thr}. tetapi tidak dapat karena sisa {total_budget}")
            continue

        total_budget -= thr
        print(f"{name}-{age}thn. dapat {thr}. sisa {total_budget}")

    print(f"sisa akhir thr: {total_budget}")


distribute_thr(total_budget=900000)
```

```python
import pandas as pd

df = pd.read_csv("../data/children_thr_data.csv")
df["age"] = df["age"].astype("int")

def distribute_thr(total_budget):
    """
    Calculate Tunjangan Hari Raya (THR) to be distributed for each of the children (given dataframe)

    total_budget: int: initial THR budget
    """
    
    thr = total_budget
    child_dapet = []
    child_gadapet = []
    
    for name,age in zip(df["name"], df["age"]):
        if age >= 1 and age <= 5:
            thr = 20000
        elif age >= 6 and age <= 10:
            thr = 30000
        elif age >= 11 and age <= 15:
            thr = 50000
        elif age >= 16 and age <= 20:
            thr = 75000
        elif age >= 21 and age <= 25:
            thr = 100000
        else:
            thr = 0

        if thr == 0:
            print(f"{name}-{age}thn tidak dapat sama sekali")
            child_gadapet.append(name)
            continue
        elif total_budget-thr < 0:
            print(f"{name}-{age}thn. harusnya dapat {thr}. tetapi tidak dapat karena sisa {total_budget}")
            child_gadapet.append(name)
            continue
        else:
            child_dapet.append(name)

        total_budget -= thr
        print(f"{name}-{age}thn. dapat {thr}. sisa {total_budget}")

    print(f"sisa akhir thr: {total_budget}")
    print(f"anak2 yg dapat thr adalah: {child_dapet}")
    print(f"anak2 yg tidak dapat thr adalah: {child_gadapet}")

distribute_thr(total_budget=900000)
```

```python
import pandas as pd

df = pd.read_csv("../data/children_thr_data.csv")
df["age"] = df["age"].astype("int")

def distribute_thr(total_budget):
    """
    Calculate Tunjangan Hari Raya (THR) to be distributed for each of the children (given dataframe)

    total_budget: int: initial THR budget
    """
    
    thr = total_budget
    child_dapet = []
    child_gadapet = []
    
    for name,age in zip(df["name"], df["age"]):
        if age >= 1 and age <= 5:
            thr = 20000
        elif age >= 6 and age <= 10:
            thr = 30000
        elif age >= 11 and age <= 15:
            thr = 50000
        elif age >= 16 and age <= 20:
            thr = 75000
        elif age >= 21 and age <= 25:
            thr = 100000
        else:
            thr = 0

        if thr == 0:
            print(f"{name}-{age}thn tidak dapat sama sekali")
            child_gadapet.append(name)
            continue
        elif total_budget-thr < 0:
            print(f"{name}-{age}thn. harusnya dapat {thr}. tetapi tidak dapat karena sisa {total_budget}")
            child_gadapet.append(name)
            continue
        else:
            child_dapet.append(name)

        total_budget -= thr
        print(f"{name}-{age}thn. dapat {thr}. sisa {total_budget}")

    print(f"sisa akhir thr: {total_budget}")
    print(f"anak2 yg dapat thr adalah: {child_dapet}")
    print(f"anak2 yg tidak dapat thr adalah: {child_gadapet}")

    return child_gadapet

child_thr = distribute_thr(total_budget=600000)

if len(child_thr) > 0:
    if any(name in child_thr for name in {"tuti", "yono"}):
        print("---selain tuti dan yono, semua anak dapat thr---")
    else:
        print("---selain tuti dan yono, tidak semua anak dapat thr---")
else:
    print("---semua anak dapat thr---")
```
