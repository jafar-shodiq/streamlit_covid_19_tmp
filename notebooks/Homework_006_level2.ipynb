```python
import pandas as pd
import numpy as np
```

```python
def distribute_thr(total_budget=900000):
    df = pd.read_csv("../data/children_thr_data.csv", sep=";")
    df = df.dropna(axis=1, how="all")
    df["age"] = df["age"].astype(float).astype(int)

    seen = set()
    hasil = []

    for name, age in zip(df["name"], df["age"]):
        if name in seen:
            continue
        seen.add(name)

        if 1 <= age <= 5:
            thr = 20000
        elif 6 <= age <= 10:
            thr = 30000
        elif 11 <= age <= 15:
            thr = 50000
        elif 16 <= age <= 20:
            thr = 75000
        elif 21 <= age <= 25:
            thr = 100000
        else:
            hasil.append(f"{name.capitalize()}-{age}thn tidak dapat sama sekali")
            continue

        if total_budget >= thr:
            total_budget -= thr
            hasil.append(f"{name.capitalize()}-{age}thn dapat Rp.{thr}. Sisa Rp.{total_budget}")
        else:
            hasil.append(
                f"{name.capitalize()}-{age}thn seharusnya dapat Rp.{thr}, "
                f"tapi tidak dapat karena sisa budget tinggal Rp.{total_budget}"
            )

    hasil.append(f"Sisa akhir THR: Rp.{total_budget}")
    return hasil

result = distribute_thr(900000)

for r in result:
    print(r)
```
